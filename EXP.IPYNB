```python
df.drop(columns=['url'],axis=1,inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df['description'].fillna('No description available', inplace=True)
df['ingredients'].fillna('No ingredients listed', inplace=True)
df['tags'].fillna('No tags', inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar_by_ingredients(recipe_index, df):
    """
    Recommend similar recipes based on ingredients of the recipe at 'recipe_index'.
    
     
    """
    df['ingredients_combined'] = df['ingredients']
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['ingredients_combined'])
    cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
    sim_scores = list(enumerate(cosine_sim[recipe_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:20]   
    recipe_indices = [i[0] for i in sim_scores]
    return df[['recipe_title', 'ingredients','cuisine', 'course', 'diet']].iloc[recipe_indices]

recommended_recipes_by_ingredients = recommend_similar_by_ingredients(4, df)
(recommended_recipes_by_ingredients)
```
